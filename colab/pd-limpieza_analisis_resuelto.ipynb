```python
def fun_result_equipo(evento, resultado):
  res_sep= resultado.split(sep=':')

  if evento == 'Home':
    return res_sep[0]

  elif evento =='Away': 
    return res_sep[1]
```

```python
for i in range(messi.shape[0]):
  messi['resultado_equipo'][i]= fun_result_equipo(messi['Evento'][i], messi['Resultado'][i])
```

ðŸ§¹ Ahora vamos con los resultados del oponente

```python
def fun_result_oponent(evento, resultado):
  res_sep= resultado.split(sep=':')

  if evento == 'Home':
    return res_sep[1]

  elif evento =='Away': 
    return res_sep[0]
```

```python
for i in range(messi.shape[0]):
  messi['resultado_oponente'][i]= fun_result_oponent(messi['Evento'][i], messi['Resultado'][i])
```

ðŸ§¹ Vamos a convertir en tipo nÃºmerico el minuto del gol

```python
messi['minuto_partido']=0
messi['minuto_extra']=0

for i in range(messi.shape[0]):
  min_sep= messi['Minuto'][i].split(sep="'")
  messi['minuto_partido'][i]= min_sep[0]
  messi['minuto_extra'][i] = min_sep[1]
```

ðŸ”Ž Â¿Mete mÃ¡s goles en primer o segundo tiempo?

![MESIJUEGO](https://media.tycsports.com/files/2022/04/19/417187/messi-meme-en-efootball-2022_w416.webp)

```python
primer_tiempo=0
segundo_tiempo=0
extra_1T=0
extra_2T=0

for i in range(messi.shape[0]):

  if messi['minuto_partido'][i] <45 :
      primer_tiempo+=1
  elif messi['minuto_partido'][i]>45 and messi['minuto_partido'][i]<90 :
      segundo_tiempo+=1
  elif messi['minuto_partido'][i]==45 and messi['minuto_extra'][i]!=0:
      extra_1T+=1
  elif messi['minuto_partido'][i]==45 and messi['minuto_extra'][i]==0:
    primer_tiempo+=1
  elif messi['minuto_partido'][i] == 90 and messi['minuto_extra'][i] == 0:
    segundo_tiempo+=1
  elif messi['minuto_partido'][i] < 90 :
    extra_2T+=1
  else:
    extra_2T+=1

print('Porcentaje de goles en el primer tiempo:', (primer_tiempo/messi.shape[0]) * 100)
print('Porcentaje de goles en el Primer tiempo extra:', (extra_1T/messi.shape[0]) * 100)
print(' ')
print('Porcentaje de goles en el Segundo tiempo:', (segundo_tiempo/messi.shape[0]) * 100)
print('Porcentaje de goles en el Segundo tiempo extra:', (extra_2T/messi.shape[0]) * 100)
```
